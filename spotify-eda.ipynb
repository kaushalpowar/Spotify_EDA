```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('../input/spotify-datasets/tracks.csv')
```

```python
df_tracks.head()
```

```python
#Null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
storted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
storted_df
```

```python
df_tracks.describe().transpose()
```

```python
most_popular = df_tracks.query('popularity>90',inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
df_tracks.set_index('release_date',inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index) #data time format
df_tracks.head()
```

```python
df_tracks[['artists']].iloc[13]
```

```python
#Changing duration from milliseconds to seconds
df_tracks['duration']= df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
#Correlation
corr_df = df_tracks.drop(['key','mode','explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap= sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1,cmap ='inferno', center=0, linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
len(sample_df)
```

```python
#Regression plot between loudness & enenrgy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness', x='energy', color='c').set(title = "Loudness vd Energy")
```

```python
#Regression plot between popularity & acousticness
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity', x='acousticness', color='g').set(title = "popularity vs acousticness")
```

```python
#Create a new column year from release date column
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
#Total no of songs in each year from 1992
#Distribution plot
sns.displot(years, discrete=True, aspect=3, height=5, kind='hist',color='orange').set(title='Number of songs per year',xlim=(1920,2021))
```

```python
total_dr = df_tracks.duration
fig_dims= (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x= years, y = total_dr, ax=ax ,errwidth=False).set(title='Year vs Duration')
plt.xticks(rotation=90)
```

```python
total_dr = df_tracks.duration
```

```python
fig_dims = (25,5)
fig, ax= plt.subplots(figsize = fig_dims)
fig=sns.pointplot(x=years, y= total_dr, ax=ax).set(title="Year vs duration")
plt.xticks(rotation=60)
```

# **Analysis on Genre of the song**

```python
df_genre = pd.read_csv('../input/ultimate-spotify-tracks-db/SpotifyFeatures.csv')
```

```python
df_genre.head()
```

```python
plt.title('Duration of th songs from Different Genre')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel('Duration in Milliseconds')
plt.ylabel("Genre of songs")
```

```python
#Top 5 genre by popularity
sns.set_style(style = 'darkgrid')
plt.figure(figsize=(10,5))
famous = df_genre.sort_values('popularity', ascending = False).head(10)
sns.barplot(y='genre', x= 'popularity', data= famous).set(title="Top 5 genre bt popularity")
```
